```python
def new_arr(x, y, z):
  '''
  to generate an array with given shape and fill value

  Params:
  param1: size of row
  param2: size of column
  param3: fill value

  return:
  array with the given shape and fill value
  '''
  return np.full((x, y), z)

arr = new_arr(2,3,7)
zeros_arr = np.zeros_like(arr, dtype=None)  # if dtype = None, it matches the parent dtype in this case, it matches with arr
ones_arr = np.ones_like(arr, dtype=None)
print(arr)
print(zeros_arr)
print(ones_arr)
```

```python
# Create a 8X8 matrix and fill it with a checkerboard pattern.
arr = np.zeros((8,8), dtype=int)
arr[1::2, ::2] = 1     # fill odd rows, even columns
arr[::2, 1::2] = 1     # fill even rows, odd columns

print(arr)
```

```python
x = np.array([[1,2],[3,5]])
y = np.array([[5,6],[7,8]])
v = np.array([9,10])
w = np.array([11,12])

print(x+y)
print(v-w)
print(5*x)
print(x**2)
print(np.dot(x,y))
print(np.dot(x,v))
print(np.dot(v,w))
print(np.vstack((x,y)))  # vstack for row wise concatenation
print(np.hstack((v,w)))  # hstack for column wise concatenation
print(np.concatenate((x,v)))  # throws a ValueError because for concatenation to work, the dimension of both the arrays must be same
# ValueError: all the input arrays must have same number of dimensions, but the array at index 0 has 2 dimension(s) and the array at index 1 has 1 dimension(s)
```

```python
A = np.array([[3,4],[7,8]])
A_inv = np.linalg.inv(A)
I = np.dot(A, A_inv)
print(I)
```

```python
A = np.array([[3,4],[7,8]])
B = np.array([[5,3],[2,1]])
result1 = np.dot(A,B)
result2 = np.dot(B,A)
print(np.array_equal(result1, result2))  # checks if the arrays are similar
```

```python
A = np.array([[3,4],[7,8]])
B = np.array([[5,3],[2,1]])
A_trans = np.linalg.matrix_transpose(A)
B_trans = np.linalg.matrix_transpose(B)
AB = np.dot(A,B)
AB_trans = np.linalg.matrix_transpose(AB)
result = np.dot(B_trans, A_trans)
print(np.array_equal(AB_trans, result))
```
